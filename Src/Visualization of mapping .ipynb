## 0. Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

## 1. Data

```python
h_45=pd.read_csv("45_hisat_sorted_coverage.txt",sep="\t")
h_46=pd.read_csv("46_hisat_sorted_coverage.txt",sep="\t")
h_47=pd.read_csv("47_hisat_sorted_coverage.txt",sep="\t")
```

```python
h_45['sample']='45'
h_46['sample']='46'
h_47['sample']='47'
```

```python
combined_data=pd.concat([h_45,h_46,h_47])
```

```python
combined_data
```

## 2.0 Coverage

```python
fig = sns.catplot(x="#rname", y="coverage", hue="sample", kind="bar", data=combined_data,height=8, aspect=15/8)

plt.xlabel("Chromosome")
plt.ylabel("Coverage")
plt.title("Mapping - coverage over chromosomes")
plt.show(fig)
```

## 3. Meandepth

```python
fig = sns.catplot(x="#rname", y="meandepth", hue="sample", kind="bar", data=combined_data,height=8, aspect=15/8)

plt.xlabel("Chromosome")
plt.ylabel("Mean depth")
plt.title("Mapping - mean depth over chromosomes")
plt.show(fig)
```

## 4. Meanbaseq

```python
fig = sns.catplot(x="#rname", y="meanbaseq", hue="sample", kind="bar", data=combined_data,height=8, aspect=15/8)

plt.xlabel("Chromosome")
plt.ylabel("Mean base quality")
plt.title("Mapping - mean base quality over chromosomes") # You can comment this line out if you don't need title
plt.show(fig)
```
